```python
# 单层装饰器
def makebold(fn):
    def wrapped():
        return "<b>" + fn() + "</b>"
    return wrapped

def makeitalic(fn):
    def wrapped():
        return "<i>" + fn() + "</i>"
    return wrapped

@makebold
@makeitalic
def hello():
    return "hello world"

hello() ## returns "<b><i>hello world</i></b>"
```

```python
# 两层装饰器,多参数
import time
import pandas as pd
def time_cost(f):
    def _f(*args):
        start = time.time()
        data = f(*args)
        end = time.time()
        print(f.__name__,"running time is",end - start)
        return data
    return _f

@time_cost
def list_comp(cols,index):
    data = {c:[str(c) + str(i) for i in index] for c in cols}  # 创建数据框
    return pd.DataFrame(data,index)

list_comp("ABCDEFGHIJKLMN",range(10000))
```
