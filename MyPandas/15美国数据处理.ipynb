```python
import numpy as np
import pandas as pd
```

```python
path = r"D:\MyDataAnalysis\MyPandas"
abbrevs = pd.read_csv(path + "\\" + "state-abbrevs.csv")
areas = pd.read_csv(path + "\\" + "state-areas.csv")
pop = pd.read_csv(path + "\\" + "state-population.csv")
```

```python
display(abbrevs.head(),areas.head(),pop.head())
```

```python
display(abbrevs.shape,areas.shape,pop.shape)
```

```python
merged = pd.merge(pop,abbrevs,how="outer",left_on="state/region",right_on="abbreviation")  # 外连接
merged.head()
```

```python
merged = merged.drop("abbreviation",axis=1)  # 删掉重复的列
merged.head(10)
```

```python
merged.isnull().any()
```

```python
merged.isnull().sum()
```

```python
merged[merged["population"].isnull()]  # 查看缺失数据
```

```python
# merged.loc[merged["state"].isnull(),"state/region"]
```

```python
merged.loc[merged["state"].isnull(),"state/region"].unique()
```

```python
# 处理缺失值
merged.loc[merged["state/region"]=="PR","state"] = "Puerto Rico"
merged.loc[merged["state/region"]=="USA","state"] = "United States"
```

```python
merged.isnull().any()
```

```python
# 左合并
final = pd.merge(merged,areas,how="left",on="state")  # 加入面积
final.shape
```

```python
final.loc[final["area (sq. mi)"].isnull(),"state/region"].unique()
```

```python
nullindex = final[final["area (sq. mi)"].isnull()].index
nullindex
```

```python
# newdata = final.drop(nullindex)
# newdata.shape
final.dropna(inplace=True)  # 删除空缺
final.shape
```

```python
final.isnull().any()
```

```python
# 查询
data2010  = final.query("year==2010 & ages=='total'")
data2010.head()
```

```python
data2010.set_index("state",inplace=True)  # 重置索引
data2010.head()
```

```python
# 计算人口密度
density = data2010["population"].div(data2010["area (sq. mi)"])
density.head()
```

```python
density.sort_values(ascending=False,inplace=True)  # 降序,本地排序
density.head(10)
```
