```python
import numpy as np
import pandas as pd
```

```python
def make_dataFrame(cols,index):
    data = {c:[str(c)+str(i) for i in index] for c in cols}
    return pd.DataFrame(data,index)
```

```python
x = make_dataFrame("ABC",[1,2])
y = make_dataFrame("BCD",[3,4])
```

```python
x
```

```python
y
```

```python
pd.concat([x,y])  # 外连接,维度不一致,并集
```

```python
pd.concat([x,y],join="inner") # 内连接,交集
```

```python
# 以x为准
pd.concat([x,y],join_axes=[x.columns])
```
